```python
url="https://www.jumia.com.eg/ar/catalog/?q=furniture&page=0#catalog-listing"
req=requests.get(url)
soup=BeautifulSoup(req.content,"html.parser")
link=soup.find_all("a",{"class":'core'})
for i in link:
    page="https://www.jumia.com.eg/"+i.attrs['href']
    print(page)
    req=requests.get(page)
    soup=BeautifulSoup(req.content,"html.parser")
    
    rate=soup.find("div",{"class":'stars _s _al'}).text
    print("rate ",rate)
    delivery=soup.find("em").text
    print("delivery ", delivery)

    try:
        sale=soup.find("span",{"class":'bdg _dsct _dyn -mls'}).text
        print("sale ",sale)
        price=soup.find("span",{"class":'-tal -gy5 -lthr -fs16'}).text
        print("price ",price)

    except:
        sale="0%"
        print("sale ",sale)
        price=soup.find("span",{"class":'-b -ltr -tal -fs24'}).text
        print("price ",price)
```

```python
name=[]
sale=[]
price=[]
page=[]
rate=[]
delivery=[]
countr=0
for x in np.arange(0,50):#all bages in site
    
    url=f"https://www.jumia.com.eg/ar/catalog/?q=furniture&page={x}#catalog-listing"
    req=requests.get(url)
    soup=BeautifulSoup(req.content,"html.parser") 
    
    nam=soup.find_all("h3",{'class':'name'}) 
    for i in nam: 
        name.append(i.text)
    
    
    link=soup.find_all("a",{"class":'core'})
    for i in link:
        pag="https://www.jumia.com.eg/"+i.attrs['href']
        page.append(pag)
        
        req=requests.get(pag)
        soup=BeautifulSoup(req.content,"html.parser")

        rat=soup.find("div",{"class":'stars _s _al'}).text
        rate.append(rat)
        
        deliver=soup.find("em").text
        delivery.append(deliver)

        try:
            sal=soup.find("span",{"class":'bdg _dsct _dyn -mls'}).text
            sale.append(sal)
            pric=soup.find("span",{"class":'-tal -gy5 -lthr -fs16'}).text
            price.append(pric)

        except:
            sal="0%"
            sale.append(sal)
            pric=soup.find("span",{"class":'-b -ltr -tal -fs24'}).text
            price.append(pric)


    countr=+1
    print(countr)
```

```python
def furniture_type(i) :
    i=re.sub("É", "E",i) 
    i=re.sub("[^A-Za-z]", " ", i)
    return i.strip()
df["type"]=df["name"].apply(furniture_type)
```

```python
df["name2"]=df["name"].apply(lambda i:re.sub("[a-zA-Z0-9]|[*]|[É]", " ", i).strip())
```

```python
from deep_translator import GoogleTranslator


def transtale_ar_en(text):
    try:
        x = GoogleTranslator(source='ar',target='en').translate(text)
    except:
        x=np.nan
    return x

df["name3"]=df["name2"].apply(transtale_ar_en)
```

```python
def cut_name(i):
    i=str(i)
    i=re.split('-',i)
    return i[0]
df["furniture"]=df["name3"].apply(cut_name)
```

```python
def fill_furniture_type(i):

    if len(i)>4 :
        return i
    else:
        return "Modern Home"
df["type"]=df["type"].apply(fill_furniture_type)
```

```python
def fill_furniture_name(i):

    if len(i)>4 :
        return i
    else:
        return np.nan
df["furniture"]=df["furniture"].apply(fill_furniture_name)
df['furniture'] = df['furniture'].fillna(df['name'])
```

```python
def to_int(i):
    try:
        i=re.sub("جنيه|,", "", i)
        return float(i)
    except:
        return np.nan
df["price"]=df["price"].apply(to_int)
df["delivery"]=df["delivery"].apply(to_int)
```
